```python
def remove_aliases(text=''):
    start_indices = [match.start() for match in re.finditer('---', text)]
    i = 0
    while i < len(start_indices):
        next_line = text.find('---', start_indices[i] + 4)
        text = text.replace(text[start_indices[i]:next_line +3], (next_line +3 - start_indices[i])* '#')
        i = i + 2
    text = text.replace('#', '')
    return text
```

```python
import re

def clean_links(text):
    start_indices = [match.start() for match in re.finditer('\[\[', text)]
    
    linktext = []
    for index in start_indices:
        if not text[index - 1] == '!':
            next_pipe = text.find('|', index)
            close_index = text.find(']', index)
            if next_pipe != -1 and next_pipe - close_index < 0:
                linktext.append(text[index+2:next_pipe].replace(" ", "_"))
                text = text.replace(text[index + 2:next_pipe + 1], (next_pipe - index -1)*'#')
            else:
                linktext.append(text[index + 2:close_index].replace(" ", "_"))
    text = text.replace('#', '')
    if '[[' in text:
        start_indices2 = [match.start() for match in re.finditer('\]\]', text)]
        i = len(linktext) - 1
        for index in reversed(start_indices2):
            if not index == len(text)-2:
                text = text[:index + 2] + '(https://defsmath.github.io/DefsMath/' + linktext[i] + ')' + text[index + 2:]
                i = i - 1
    text = text.replace('[[', '[')
    text = text.replace(']]', ']')
    
    return text
```

```python
def do_wikidatas(text):
    if 'https://www.wikidata.org/wiki/Q' in text:
        index = text.index('https://www.wikidata.org/wiki/Q')
        text = text[:index] + 'Wikidata ID: [' + text[index+30:].strip() + '](https://www.wikidata.org/wiki/' + text[index+30:].strip() + ')'
    return text
```

```python
text = open('/Users/lucyhorowitz/Documents/GitHub/DefsMath/definitions/adjoint of a linear transformation.md').read()
text = remove_aliases(text)
text = clean_links(text)
text = do_wikidatas(text)
print(text)
```
